```python
import torch
import pandas as pd
```

```python
if torch.cuda.is_available():
  device = torch.device('cuda')
else:
  device = torch.device('cpu')
print('Device', device)
```

```python
# Load the CSV file
input_file = 'output_vectors/VGGNet_Total_output_vectors_training.csv'
df = pd.read_csv(input_file)

# Ensure the data is sorted by date
df = df.sort_values(by=['date'])

# Group by date and create a list of stocks sorted by their vectors for each date
output_data = []
for date, group in df.groupby('date'):
    # Sort the group by vector in descending order
    sorted_group = group.sort_values(by='vector', ascending=False)
    # Create a list of stocks sorted by their vectors
    sorted_stocks = sorted_group['stock'].tolist()
    # Append the date and sorted stocks to the output data
    output_data.append({'Date': date, 'Ranking': sorted_stocks})

# Create a DataFrame from the output data
output_df = pd.DataFrame(output_data)
print(output_df)

# Save the output DataFrame to a new CSV file
output_df.to_csv('data_for_strategy_testing/VGGNet_Total_rank_training.csv', index=True)
print('completed')
```
